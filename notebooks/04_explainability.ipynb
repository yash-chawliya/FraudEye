# Phase 7: Explainability (SHAP)

1. Compute SHAP values for the trained XGBoost model
2. Visualize top contributing features

```python
import shap
import xgboost as xgb
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def explain_model(model, X_train):
    # Initialize SHAP explainer
    explainer = shap.TreeExplainer(model)
    # Compute SHAP values for a subset (due to large dataset)
    shap_values = explainer.shap_values(X_train.sample(1000, random_state=42))
    
    # Summary plot
    shap.summary_plot(shap_values, X_train.sample(1000, random_state=42))
    
    # Bar plot for global feature importance
    shap.summary_plot(shap_values, X_train.sample(1000, random_state=42), plot_type="bar")
```
